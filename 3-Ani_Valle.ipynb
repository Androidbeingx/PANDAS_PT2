### Import Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=UserWarning)


data = pd.read_csv("./data/tidy_seer_breast_cancer_dataset.csv")
data.head(5)
```

## Question 1.

    Displays a graph showing the size of the tumor based on the grade.

```python
# Chart Size
plt.figure(figsize=(15,8))

# Calculate the data
sns.boxplot(x='tumor_size', y='grade', data=data)

# Chart labels
plt.xlabel('Tumor Size',fontdict={'fontsize':15})
plt.ylabel('Grade',fontdict={'fontsize':15})
plt.title("Size of the Tumour based on Grade",fontdict={'fontsize':20})
```

---

### _CONCLUSION_

Patients with grade 4 breast cancer have a larger tumor compared to grades 1, 2, and 3.

Since grade 4 breast cancer cancer spreads to distant organs.
For example, the lungs or bones.

## Question 2.

    Estrogen and progesterone status according to the grade of breast cancer.

```python
fig, axis = plt.subplots(1, 2, figsize=(18,8))

sns.countplot(data.grade, hue=data['estrogen_status'], ax=axis[0])
axis[0].set_title("Estrogen status for different Grade of Breast Cancer", fontweight="bold", fontdict={'fontsize':15})

sns.countplot(data.grade, hue=data['progesterone_status'], ax=axis[1])
axis[1].set_title("Progesterone status for different Grade of Breast Cancer", fontweight="bold", fontdict={'fontsize': 15})
```

---

### _CONCLUSION_

The graph shows that the state of Estrogen and Progesterone are mostly positive.

What does this mean?
Breast cancer can be "hormone receptor positive" or HR+, if it does not have hormone receptors it is hormone receptor negative or HR.

If the tumor is HR+, the tumor needs estrogen and/or progesterone to grow.

As we can see in "grade 2" cancer we have that the estrogen and progesterone status is mostly positive, compared to grade 3 and 4.
This may be because in some cancer treatments HR+ can be treated with endocrine therapy (also called hormone therapy) to block the action of hormones.

In early-stage breast cancer, hormonal therapies may be used along with chemotherapy and targeted therapy.

But this does not ensure that the tumor will shrink or stop advancing.

With the previous graph we saw that the size of the tumor is greater in grade 4, so in the early stages this hormonal treatment could be used and stop the production of these hormones, but even so the tumor could not stop growing.

## Question 3.

    Survival probability.
        - According to the number of months that the tumor was in the body.
        - Depending on the grade of cancer
        - Depending on progesterone and estrogen status.

__3.1 According to the number of months that the tumor was in the body__

```python
plt.figure(figsize=(40,25))
ax = sns.countplot(data['survival_months'], hue=data['status'])

plt.xlabel("survival_months",fontdict={'fontsize': 15})
plt.ylabel("Count",fontdict={'fontsize': 15})
plt.title("Patient dead or alive based on Survival Months",fontdict={'fontsize': 20}, fontweight="bold")
```

---

### CONCLUSION

Most of the people who lived had the tumor in their body for more months.

---

__3.2 Depending on the grade of cancer__

```python
plt.figure(figsize=(15,8))
ax =sns.countplot(data=data, x="grade", hue="status")

for p in ax.patches:
   ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))

plt.xticks(rotation = 30)
plt.xlabel("Grade",fontdict={'fontsize': 15})
plt.ylabel("Count",fontdict={'fontsize': 15})
plt.title("Patient dead or alive based on Grade",fontdict={'fontsize': 20}, fontweight="bold")
```

---

### CONCLUSION

There are a greater number of people who lived with breast cancer in "grade 2".

---

__3.3 Según la progesterona y el estrogeno__

```python
fig, axis = plt.subplots(1, 2, figsize=(18,8))

sns.countplot(data.status, hue=data['estrogen_status'], ax=axis[0])
axis[0].set_title("Patient dead or alive based on Estrogen status", fontweight="bold", fontdict={'fontsize':15})

sns.countplot(data.status, hue=data['progesterone_status'], ax=axis[1])
axis[1].set_title("Patient dead or alive based on Progesterone status", fontweight="bold", fontdict={'fontsize': 15})
```

---

### CONCLUSION

Most people who were positive for estrogen and progesterone status survived.
This may be because "grade 2" is the one that is usually controlled and treated the most, since people are still physiologically strong to receive treatment.
