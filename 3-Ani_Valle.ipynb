>### IMPORTS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

>### GET DATA

```python
data = pd.read_csv("./data/tidy_seer_breast_cancer_dataset.csv")
data
```

>## Question 1.

Displays a graph showing the size of the tumor based on the grade.

```python
# Chart Size
plt.figure(figsize=(15,8))

# Calculate the data
sns.boxplot('tumor_size', 'grade', data=data)

# Chart labels
plt.xlabel('Tumor Size',fontdict={'fontsize':15})
plt.ylabel('Grade',fontdict={'fontsize':15})
plt.title("Size of the Tumour based on Grade",fontdict={'fontsize':20})
```

>### CONCLUSION

Patients with grade 4 breast cancer have a larger tumor compared to grades 1, 2, and 3.
Since grade 4 breast cancer cancer spreads to distant organs.
For example, the lungs or bones.

>## Question 2.

Estrogen and progesterone status according to the grade of breast cancer.

```python
fig, axis = plt.subplots(1, 2, figsize=(16,8))

sns.countplot(data.grade, hue=data['estrogen_status'], ax=axis[0])
axis[0].set_title("Estrogen status for different Grade of Breast Cancer", fontweight="bold", fontdict={'fontsize':15})

sns.countplot(data.grade, hue=data['progesterone_status'], ax=axis[1])
axis[1].set_title("Progesterone status for different Grade of Breast Cancer", fontweight="bold", fontdict={'fontsize': 15})
```

>### CONCLUSION

The graph shows that the state of Estrogen and Progesterone are mostly positive.

What does this mean?
Breast cancer can be "hormone receptor positive" or HR+, if it does not have hormone receptors it is hormone receptor negative or HR.

If the tumor is HR+, the tumor needs estrogen and/or progesterone to grow.

As we can see in "grade 2" cancer we have that the estrogen and progesterone status is mostly positive, compared to grade 3 and 4.
This may be because in some cancer treatments HR+ can be treated with endocrine therapy (also called hormone therapy) to block the action of hormones.

In early-stage breast cancer, hormonal therapies may be used along with chemotherapy and targeted therapy.

But this does not ensure that the tumor will shrink or stop advancing.

With the previous graph we saw that the size of the tumor is greater in grade 4, so in the early stages this hormonal treatment could be used and stop the production of these hormones, but even so the tumor could not stop growing.
